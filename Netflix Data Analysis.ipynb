### Q2: Which year has the highest number of TV Shows and Movies released? Show with a bar graph.

```python
# Making a new column which makes the data type for release data datetime instead of object
```

```python
df['Date_N'] = pd.to_datetime(df['Release_Date'])
```

```python
df.head() 
```

```python
df['Date_N'].dt.year.value_counts() #counts the number of times a year is in the date column
```

```python
# Making the bar graph
df['Date_N'].dt.year.value_counts().plot(kind= 'bar',title= 'TV Shows and Movies Released Per Year', xlabel='Year', ylabel='Count')
```
